```python
import pandas as pd

# Read the comma-separated CSV properly
messages = pd.read_csv('spam.csv', encoding='latin-1')

# Rename relevant columns if needed
messages = messages.rename(columns={
    messages.columns[0]: "label",
    messages.columns[1]: "message"
})

# Drop unnecessary columns if they exist (depends on your CSV)
messages = messages[["label", "message"]]

# Check the result
messages.head()
```

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()       
    review=[lemmatizer.lemmatize(word) for word in review ]
    review=' '.join(review)     
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
import gensim
##lets train word2vec model
model=gensim.models.Word2Vec(words)
```
